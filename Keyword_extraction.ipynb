```python
import spacy
import pandas as pd
import numpy as np
```

```python
nlp = spacy.load("en_core_web_sm")

def get_keywords(text):
  keywords = []
  doc = nlp(text)

  for token in doc:
      # print(token)
      if (token.pos_ == 'NOUN'):
        
        keywords.append(token.text)

  return keywords
```

```python
link = 'https://drive.google.com/uc?export=download&id=1yHGTAm0IyyUUHUchv6aAdBqYOdAGWHPu'
df = pd.read_csv(link)
df.head() 
```
