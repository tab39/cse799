```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
import seaborn as sns

%matplotlib inline
```

## Generate heatmap

```python
def generateHeatmap(df, key, value, pivots, figsize):
  country = df.loc[df[key] == value]
  country = country.pivot(pivots[0], pivots[1], 'count')

  f, ax = plt.subplots(figsize=figsize)
  sns.heatmap(country, annot=True, ax=ax, cmap="Spectral_r", linewidths=.005, linecolor='black')
```

--------------------------------------------------

## Pivot 1: Attack organisation - attacked country (organised by attacking country / entity category)

--------------------------------------------------

```python
df = pd.read_csv('by_attack_organisation.csv')
```

```python
attackingCountries = df['attacking_country'].unique()
print(attackingCountries)
```

```python
def getHeatmap1(key, value, figsize):
  pivots = ['attacking_organisation', 'attacked_country']
  generateHeatmap(df, key, value, pivots, figsize)
```

### Belarus

```python
figsize = (2.5, 1)
getHeatmap1('attacking_country', 'Belarus', figsize)
```
